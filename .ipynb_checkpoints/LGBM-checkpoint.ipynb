# Moving Average Crossover Strategy

## Importing Data
### Stock: NVDA \ Period: 1Y, Interval: 1D


```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
nvda = yf.Ticker("NVDA")

data = nvda.history(period = "1y", interval = "1d")

plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label='Nvidia Closing Price')
plt.title('Nvidia (NVDA) Closing Price - Past 1 Year')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## LGBM Implementation
### Utilize Libraries

```python
# --- Imports ---
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
import numpy as np

# --- Feature Engineering ---
data['Return'] = data['Close'].pct_change()
data['Momentum'] = data['Close'] - data['Close'].shift(1)
data['Std5'] = data['Close'].rolling(5).std()
data['HighLow5'] = data['Close'].rolling(5).max() - data['Close'].rolling(5).min()
data['RollingMean5'] = data['Close'].rolling(5).mean()

# Lagged returns for context
for lag in range(1, 11):
    data[f'Return_lag{lag}'] = data['Return'].shift(lag)

# Target: 1 if next day close > today, else 0
data['Target'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Drop NA values
data.dropna(inplace=True)

# Features and target
features = ['Close', 'Return', 'Momentum', 'Std5', 'HighLow5', 'RollingMean5'] + [f'Return_lag{i}' for i in range(1,11)]
X = data[features]
y = data['Target']

# --- Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# --- Check class balance ---
print("Class distribution in training set:")
print(y_train.value_counts(normalize=True))

# --- Prepare LightGBM datasets ---
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# --- LGBM Parameters ---
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'verbose': -1,
    'scale_pos_weight': (y_train==0).sum() / (y_train==1).sum()  # handle imbalance
}

# --- Train Model with Early Stopping ---
model = lgb.train(
    params,
    train_data,
    valid_sets=[test_data],
    num_boost_round=200,
    callbacks=[lgb.early_stopping(stopping_rounds=20)]
)

# --- Generate Buy/Sell Signals using median threshold ---
data['Pred'] = model.predict(X)
threshold = np.median(data['Pred'])
data['Signal'] = -1  # default sell
data.loc[data['Pred'] > threshold, 'Signal'] = 1  # buy

# --- Simulate Trades ---
position = 0
buy_price = 0
revenue = 0
volume = 100

for i in range(len(data)):
    signal = data['Signal'].iloc[i]
    price = data['Close'].iloc[i]

    if signal == 1 and position == 0:
        buy_price = price
        position = 1
        print(f"BUY at {buy_price:.2f} on {data.index[i].date()}")
    elif signal == -1 and position == 1:
        sell_price = price
        trade_revenue = (sell_price - buy_price) * volume
        revenue += trade_revenue
        position = 0
        print(f"SELL at {sell_price:.2f} on {data.index[i].date()}, Trade Revenue: {trade_revenue:.2f}")

# Close any open position at the end
if position == 1:
    sell_price = data['Close'].iloc[-1]
    trade_revenue = (sell_price - buy_price) * volume
    revenue += trade_revenue
    print(f"Final SELL at {sell_price:.2f} on {data.index[-1].date()}, Trade Revenue: {trade_revenue:.2f}")

print(f"\nTotal Revenue from LGBM strategy with volume {volume}: ${revenue:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))

# Plot close price
plt.plot(data.index, data['Close'], label='Close', color='blue')

# Mark Buy signals
buy_signals = data[data['Signal'] == 1]
plt.scatter(buy_signals.index, buy_signals['Close'], marker='^', color='green', s=100, label='Buy Signal')

# Mark Sell signals
sell_signals = data[data['Signal'] == -1]
plt.scatter(sell_signals.index, sell_signals['Close'], marker='v', color='red', s=100, label='Sell Signal')

# Optional: Shade predicted up/down regions
plt.fill_between(data.index, data['Close'], data['Close'].min(),
                 where=(data['Signal'] == 1),
                 color='green', alpha=0.1)
plt.fill_between(data.index, data['Close'], data['Close'].min(),
                 where=(data['Signal'] == -1),
                 color='red', alpha=0.1)

plt.title("LGBM Trading Strategy Signals")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
